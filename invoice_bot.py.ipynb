```python
# Project: Invoice Automation Bot
# Author: Najmuddin Saiyed
# Description: Extracts total amount from a PDF invoice using PyPDF2

import PyPDF2


try:
    file = open('invoice_101.pdf', 'rb')
    
    # 2. Create the Reader
    reader = PyPDF2.PdfReader(file)
    page = reader.pages[0]
    
    # 3. Extract Raw Text
    text = page.extract_text()
    
    # 4. Parse the Amount
    # Splitting at '$' to isolate the value
    parts = text.split('$')
    if len(parts) > 1:
        amount = parts[1].strip()
        print(f"✅ Success! Extracted Amount: ${amount}")
    else:
        print("⚠️ No amount found.")

except FileNotFoundError:
    print("❌ Error: 'invoice_101.pdf' not found. Please generate it first.")
```
